## 데이터 시각화
    - 데이터 분석 결과의 시각적 표현
    - 도표를 사용한 명확하고 효과적인 정보 전달 기술

- 데이터 시각화 절차의 단계 분류
    1. 정보 조직화 단계
    - 사용자의 정보 인지에 관여
    - 혼돈의 상태로 존재하는 데이터를 분류하고 배열하고 조직화하여 질서를 부여

    2. 정보 시각화 단계
    - 사용자의 정보 지각에 관여
    - 보다 효율적으로 정보 전달을 위해 시각, 청각, 촉각, 미각, 후각의 감각 기관에 최적의 자극을 제시하는 방법을 제시

    3. 상호작용 단계
    - 정보와 사용자 간의 상호작용 측면의 사용자 경험을 디자인
    - 정보의 이인지적 요인뿐만 아니라 지각적 요인을 함께 활용
    - 정보 시각화 단계와 밀접하게 연동되면서 동시에 입력 기술의 특성도 함께 고려
<br>
- 연관 용어
    - 정보 시각화: 빅데이터를 색체, 통계, 이미지 등을 활용해 요약적으로 표현
    - 과학적 시각화: 실험 결과나 시뮬레이션 데이터 등 복잡한 데이터를 쉽게 탐색할 수 있도록 3차원 그래픽 기술 등을 활용해 시각화 (정보 시각화의 일부)
    - 인포그래픽: 복잡한 수치나 글로 표현되어 있는 다양한 정보를 차트, 지도, 다이어그램 등을 활용해 한눈에 파악할 수 있도록 하는 디자인

````python
import matplotlib.pyplot as plt
import numpy as np # 수학 함수 사용하는 라이브러리

# 여러 개의 그래프 그리기 
x = np.arange(0, 10, 2) # 배열화 0 ~ 9 중 2개씩 짝수
y = np.arange(1, 11, 2) # 홀수
plt.plot(x, y, x, y*2, x, y**2, x, y^2)
plt.show()
````

````python
# 모듈 import 하기
import matplotlib.pyplot as plt

# 수식 y = x ^ 2 그리기
# x 범위 정하기
x = range(0, 100)

# y 값 구하기
y = [v*v for v in x] # x 번 동안 v*v

# 2차원 (x, y) 그래프 지정
plt.plot(x, y)
# 그래프 보이기
plt.show()
````

> 히스토그램
- 도수분포표를 그래프로 나타낸 것
- 누적값
- 가로축은 계급, 세로축은 도수 (횟수나 개수 등)

>> 기본 히스토그램

````python
weight = [1,1,2,3,4,5]
plt.hist(weight) # default
plt.hist(weight, orientation='horizontal') # 수평 히스토그랭
plt.show()
````

>> 구간 개수 지정

````python
plt.hist(weight, bins = 10, label = 'bins=10')
plt.legend()
plt.show()
````

>> 누적 히스토그램

````python
plt.hist(weight, cumulative = True, label='cumulative=True') # 누적 해서 보여줌
plt.hist(weight, cumulative = False, label='cumulative=False')
plt.legend(loc='upper left')
plt.show()
````

````python
plt.hist(weight, histtype='step') # 등 옵션으로 지정
plt.show()
````
